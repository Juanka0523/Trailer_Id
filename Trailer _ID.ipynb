```python
data_raw = pd.read_csv("https://raw.githubusercontent.com/jrartd/data-analytics/master/867162027426394_data_Trailer_ID.csv",
                       sep=";",header=None, parse_dates=[1])
data_raw.head(5)
```

# Extraer respuesta sin parser

```python
GTIDA = data_raw[data_raw["Sentence"].str.contains(":GTIDA")]
GTIDA
```

```python
GTDIS = data_raw[data_raw["Sentence"].str.contains(":GTDIS")]
GTDIS
```
